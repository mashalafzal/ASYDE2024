```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the dataset
input_file = "path to your train split file"
df = pd.read_csv(input_file)
```

```python
# Count the number of occurrences of each label
label_counts = df['label'].value_counts()
```

```python
# Display the shape of the dataframe
print(f"Shape of the dataframe: {df.shape}")

# Display the first few rows of the dataframe to confirm headers
df.head()
```

```python
# Create a bar chart with total numbers on top of each bar
plt.figure(figsize=(8, 6))
bars = plt.bar(label_counts.index, label_counts.values, color=['blue', 'orange'])

# Add text annotations on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height), 
                 xytext=(0, 2),  # 2 points vertical offset
                 textcoords="offset points", 
                 ha='center', va='bottom', 
                 fontsize=12, color='black', weight='bold')

plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution of Labels in the Train set')
plt.xticks(ticks=[0, 1], labels=['0', '1'])
plt.show()
```
